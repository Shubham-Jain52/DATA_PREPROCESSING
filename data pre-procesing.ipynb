## Chapter 2: How to deal with missing values in a dataset


there are two types of missing value:
1) the standard type: here all the missing values are represented by NaN or null values by default. these can be identified by the pandas isNull command.

2) The non-standard type: here the missing value may be replaced by some sort of string such as "?", "--", etc. These values can not be identified by the pandas and are not considered as null values. therefore these need to be replaced by some null values first and then those values can be replaced. 

```python
#  to replace the non standard values we use the .replace command
df1.replace("?",np.nan,inplace=True);
df1.head()

# here in this dataset we had a non standard missing value that was "?" and we first replaced it with np.nan 
```

now to identify these null values we have two methods:
1) .isNull()
2) .notnull()

```python
# to get a summary of the total number of null values in each columns.
df1.isnull().sum()
```

### Dealing methods for null values:

to deal with null values we have 3 techniques:

1) if the null values is in a numercial columns just replace all the values with mean of the column.
2) if in a column a the majority of the values are null just drop that whole column or row.
3) if nothing can be done just leave the values like that.


```python
# Method 1:
df1.dtypes

# we see that normalized-losses has numerical values and also null values but is of datatype 'object' therefore we first convert the data type.
df1_mean=df1['normalized-losses'].astype('float').mean()
print(df1_mean)

# now that we have calculated the mean of the column we can replace all the null values with the mean.
df1['normalized-losses'].replace(np.nan,df1_mean,inplace=True)
```

```python
df1.dtypes
```

```python
# similarly  for all the other numerical columns first convert their datatype if needed and then replace their null values with the mean.
df1_bore_mean=df1['bore'].astype('float').mean()
df1_stroke_mean=df1['stroke'].astype('float').mean()
df1_hp_mean=df1['horsepower'].astype('float').mean()
df1_rpm_mean=df1['peak-rpm'].astype('float').mean()
df1['bore'].replace(np.nan,df1_bore_mean,inplace=True)
df1['stroke'].replace(np.nan,df1_stroke_mean,inplace=True)
df1['horsepower'].replace(np.nan,df1_hp_mean,inplace=True)
df1['peak-rpm'].replace(np.nan,df1_rpm_mean,inplace=True)
```

```python
# for the door columns we can't just replace the doors with mean values as it either has 2 or 4 doors there fore we replace the doors with the doors with highest frequency.

df1['num-of-doors'].value_counts()

# we see that we have more number of cars with four doors. Hence we replace null with four doors.
```

```python
df1['num-of-doors'].replace(np.nan,'four',inplace=True)
```

```python
# for the price column we can't edit the prices of cars with null values with mean as the model will need the exact values only for better predictions hence we will just drop them as it is.

df1.dropna(subset='price',inplace=True,axis=0)


# since we have deleted the rows therefore we need to reset the indexes.
df1.reset_index(drop=True,inplace=True)
df1.info()
```

the strategy can be:
1) 'mean'

4) 'constant'

## Chapter 4 : Data formating and data binning 

#### DATA BINNING:
it is a method for creating bins for seprating data of a column into 3 frequency namely 'low', 'mid', and 'high'.

for example in the horsepower column this method is requried to find out how many of cars have low HP, mid HP, high HP which is crutical for categorising the cars based on the HP.

```python
# for dividing the data into 3 categories we need the range of the data seprated into low mid high, for that we need to divide the data into 3 equal parts by using a numpy command:

bins_range=np.linspace(min(df1['horsepower']), max(df1['horsepower']), 4)  # 4 cuz that break it into 3 equal parts. we also need the min and max of that column.

bins_range
```

Now we can divide the dataset into low (48 - 119.33333), mid (119.3333 - 190.6666666), high (190.66666 - 262).

#### Creating bins:

```python
bin_names = ['Low', 'Mid', 'High']

df1['horsepower_bins']=pd.cut(df1['horsepower'], bins_range, labels=bin_names, include_lowest=True)

df1[['horsepower_bins', 'horsepower']].head(10)
```

to find the frequency of these low, mid, high:

```python
df1['horsepower_bins'].value_counts()
```

Hence we find that the Low HP category has the highest amount of cars.

```python
import matplotlib.pyplot as plt
plt.bar(bin_names,df1['horsepower_bins'].value_counts())
plt.xlabel('Horsepower')
plt.ylabel('Frequency')
```
